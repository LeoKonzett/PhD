```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
dir = r"C:\Users\SLAB2\Downloads"
f_name = "pid_tf_test.txt"
fp = os.path.join(dir, f_name)
data_pid = np.loadtxt(fp, comments="%", delimiter=";", skiprows=5)
```

```python
k_p = -1.745
k_i = -1.012

data_x = np.linspace(data_pid[0, 0], data_pid[-1, 0], 1_000)
data_y = (k_p + k_i / data_x) 
```

```python
plt.plot(data_pid[:, 0], data_pid[:, 1] ** 2, label="PID TF ZI")

plt.plot(data_x, np.sqrt(data_y ** 2), label="TF: P + I/s")
plt.yscale("log")
plt.xscale("log")
plt.grid()
plt.legend()
```
