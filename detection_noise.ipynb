```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
dir = r"C:\Users\SLAB2\Documents\Leo_Thesis_Data\interferometer_characterization\detection_noise"
f_electronics = "noise_PDandLockIn_5kHzSpan_2_5kHzBW_wFilterCompensation.txt"
fp_elctr = os.path.join(dir, f_electronics)

f_laser = "noise_ORION_5kHzSpan_2_5kHzBW_ref50uW.txt"
fp_laser = os.path.join(dir, f_laser)

f_total = "noise_total_5kHzSpan_2_5kHzBW_ref50uW.txt"
fp_total = os.path.join(dir, f_total)
```

```python
data_elctr = np.loadtxt(fp_elctr, comments="%", delimiter=";", skiprows=5)
data_laser = np.loadtxt(fp_laser, comments="%", delimiter=";", skiprows=5)
data_total = np.loadtxt(fp_total, comments="%", delimiter=";", skiprows=5)
```

```python
plt.plot(data_elctr[:,0] * 1e-3, np.sqrt(data_elctr[:,1]), label="Electronics")
plt.plot(data_laser[:,0] * 1e-3, np.sqrt(data_laser[:,1]), label="Laser")
plt.plot(data_total[:,0] * 1e-3, np.sqrt(data_total[:,1]), label="Total")
plt.grid()
plt.xlabel("Frequency [kHz]")
plt.yscale("log")
plt.ylabel("PSD [V/root-Hz]")
plt.title("Noise referenced to 55uW @ PD. \n 2.5kHz demodulation BW. ")
plt.hlines(y=np.sqrt(3.5e-15), xmin=data_elctr[0, 0]*1e-3, xmax=data_elctr[-1, 0]*1e-3, linestyle="--", label="Noise floor: 5.9 E-8 [V/root-Hz]")
plt.ylim(2e-8, 1e-5)
plt.legend()
print(np.sqrt(3.5e-15))
```
